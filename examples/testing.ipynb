```python
import marvin
from marvin.infra.chroma import Chroma
from marvin.loaders.github import GitHubRepoLoader

marvin.settings.log_level = 'DEBUG'

loader = GitHubRepoLoader("prefecthq/prefect", "**/*.md", "**/docs/api-ref/**")

digest = await loader.load()

chroma = Chroma()

await chroma.add(
    **digest.dict(),
)

chroma.client.persist()
```

```python

import marvin
from marvin.bots import Bot
from marvin.plugins.chroma import SearchChromaEmbeddings

marvin.settings.log_level = 'DEBUG'

bot = Bot(name='Marvin', personality='Marvin from Hitchhikers Guide to the Galaxy', plugins=[SearchChromaEmbeddings()])

await bot.say('what CLI command do I need to make a Prefect deployment? use chroma to find out')
```

```python
results = await chroma.query(
    query_texts=["How do I run subflows on their own infrastructure?"],
    n_results=3,
    include=["metadatas", "documents"],
)

summary_text = "\n".join(
    f"{idx}\n{doc}\n" for idx, doc in zip(results["ids"], results["documents"])
)

print(summary_text)
```
